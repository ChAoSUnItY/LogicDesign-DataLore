```kotlin
operator fun <T> List<T>.component6() = this[5]
operator fun <T> List<T>.component7() = this[6]
operator fun <T> List<T>.component8() = this[7]
operator fun <T> List<T>.component9() = this[8]
operator fun <T> List<T>.component10() = this[9]
operator fun <T> List<T>.component11() = this[10]

fun List<Boolean>.toIntList() =
    this.map { b -> if (b) 1 else 0 }.toList()

fun <T> List<List<T>>.transpose(): List<List<T>> =
    List(this[0].size) { j ->
        List(this.size) { i -> 
            this[i][j]
        }
    }

fun pers(count: Int): MutableList<MutableList<Boolean>> {
    val columns = 2.0.pow(count).toInt()
    val permutations = mutableListOf<MutableList<Boolean>>()
    var num = 0

    for (i in 0 until columns) {
        val current = MutableList<Boolean>(count) { _ -> false }

        for (j in 0 until count) {
            current[count - j - 1] = ((num shr j) and 1) != 0
        }

        permutations += current
        num++
    }

    return permutations
}

fun per(): List<Boolean> =
    listOf(false, true)

fun or(vararg b: Boolean): Boolean =
    b.reduce(Boolean::or)

fun and(vararg b: Boolean): Boolean =
    b.reduce(Boolean::or)

fun nor(vararg b: Boolean): Boolean =
    b.reduce { l, r -> l.or(r).not() }

fun nand(vararg b: Boolean): Boolean =
    b.reduce { l, r -> l.and(r).not() }

fun xor(vararg b: Boolean): Boolean =
    b.reduce(Boolean::xor)
```

```kotlin
val inputs = arrayOf("M", "A4" ,"A3", "A2", "A1", "B4", "B3", "B2", "B1", "C0", "S4", "S3", "S2", "S1")
val permutations = pers(9)

fun fullAdd(a: Boolean, b: Boolean, c: Boolean) =
    ((a xor b) xor c) to ((a and b) or (b and c) or (a and c))

fun rippleCarryAdd(a: Array<Boolean>, b: Array<Boolean>, m: Boolean): Pair<List<Boolean>, Boolean> {
    var c = m
    val sl = mutableListOf<Boolean>()
    val tb = b.map { it xor c }

    for (i in 0 until 4) {
        val (s, tc) = fullAdd(a[3 - i], tb[3 - i], c)
        c = tc
        sl += s
    }

    return sl to c
}

for ((i, perm) in permutations.withIndex()) {
    val (m, a4, a3, a2, a1, b4, b3, b2, b1) = perm
    val (s, c) = rippleCarryAdd(arrayOf(a4, a3, a2, a1), arrayOf(b4, b3, b2, b1), m)

    permutations[i] += c
    permutations[i] += s
}

val df = permutations.transpose().mapIndexed { i, it -> 
    it.toIntList().toColumnOf<Number?>(inputs[i])
}.toDataFrame()

val file = File("Lab07.csv")

if (!file.exists()) {
    file.createNewFile()
}

df.writeCSV(file)

df
```
